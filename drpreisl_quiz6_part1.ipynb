```python
def reward_print(reward_per_episode, info, episodes = 1000): 
    mins = int(min(reward_per_episode)) * (.8)
    maxs = int(max(reward_per_episode)) * (1.2)
    plt.figure()
    plt.plot(reward_per_episode)
    plt.xlabel('Step', fontsize=20)
    plt.ylabel('Reward', fontsize=20)
    plt.title(f'Reward per step for {info}', fontsize=24)
    plt.xticks(fontsize=18)
    plt.yticks(fontsize=18)
    plt.ylim(ymin= mins, ymax=maxs)
    plt.xlim(xmin=0, xmax=episodes)
    plt.grid()
    plt.show()
```

```python
#Creates a random enviroment for the cart pole system and prints the reward graph
cart = gym.make("CartPole-v1", render_mode="human", max_episode_steps= 1001)
observation, info = cart.reset(seed=42)
total_cart_rewards = []
for _ in range(1001):
   action = cart.action_space.sample()  # this is where you would insert your policy
   observation, reward, terminated, truncated, info = cart.step(action)
   total_cart_rewards.append(reward)
   if terminated or truncated:
      observation, info = cart.reset()

cart.close()
reward_print(total_cart_rewards, "Cart Pole")
```

```python
car = gym.make("MountainCar-v0", render_mode="human", max_episode_steps= 1001)
observation, info = car.reset(seed=42)
total_car_rewards = []
for _ in range(1001):
   action = car.action_space.sample()  # this is where you would insert your policy
   observation, reward, terminated, truncated, info = car.step(action)
   total_car_rewards.append(reward)
   if terminated or truncated:
      observation, info = car.reset()

car.close()
reward_print(total_cart_rewards, "Mountain Car")
```
